```python
class NodoDoble:
  def __init__( self , value , siguiente=None , previo=None):
    self.data = value
    self.next = siguiente
    self.prev = previo
```

```python
class DoubleLinkedList:
    def __init__( self  ):
        self.head = None
        self.tail = None
        self.size = 0

    def get_size( self ):
        return self.size

    def is_empty( self ): # true si la lista está vacía
        return self.size == 0

    def append( self , value ):
        if self.is_empty():
            self.head = NodoDoble( value)
            self.tail = self.head
        else:
            self.tail.next = NodoDoble(value , None , self.tail )
            self.tail = self.tail.next
        self.size += 1

    def find_for_head ( self , value ):
      curr_node = self.head
      aux = False
      while curr_node != None :
          if curr_node.data == value:
            print("Se encontró el nodo")
            aux = True
          curr_node = curr_node.next
      print(aux)

    def find_for_tail ( self , value ):
      curr_node = self.tail
      aux = False
      while curr_node != None :
          if curr_node.data == value:
            print("Se encontró el nodo")
            aux = True
          curr_node = curr_node.prev
      print(aux)

    def transversal( self ):
        curr_node = self.head
        while curr_node != None :
            print(f"| {curr_node.data} | --> " , end="")
            curr_node = curr_node.next
        print("")

    def reverse_transversal ( self ):
      curr_node = self.tail
      while curr_node != None :
          print(f"| {curr_node.data} | --> " , end="")
          curr_node = curr_node.prev
      print("")

    def remove_from_head( self , value  ):
        curr_node = self.head
        while curr_node.data != value:
            curr_node = curr_node.next

        if curr_node.data == value:
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev
        self.size -= 1

    def remove_from_tail ( self , value ):
      curr_node = self.tail
      while curr_node.data != value:
          curr_node = curr_node.prev
      
      aux = curr_node.next
      aux1 = curr_node.prev
      if curr_node.data == value:
          curr_node.prev.next = aux
          curr_node = aux1
      self.size -= 1
```

```python
ldl = DoubleLinkedList()
ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.transversal()
print("\nReverse_transversal:")
ldl.reverse_transversal()
print("\nDespues de is_empty:")
print(f"Esta vacia?: {ldl.is_empty() }")
print("\nDespues de remove_from_head(20):")
ldl.remove_from_head(20)
ldl.transversal()
print("\nDespues de find_for_ head(30):")
ldl.find_for_head(30)
print("\nDespues de fin_for_tail(30):")
ldl.find_for_tail(30)
print("\nDespues de remove_for_tail(30):")
ldl.remove_from_tail(30)
ldl.transversal()
```
